```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup 
```

```python
# crawling 2020 hitter

def get_hitter(url) :
    response = requests.get(url) 
    html = response.text 
    soup = BeautifulSoup(html, "html.parser")
    soup.get('span')
    score=[i.get_text() for i in soup.select('span')]

    team_score = [[],[],[],[],[],[],[],[],[],[]] ; start = [] ; end = []
    start.append(73) ; end.append(92) ; team_score[0] = score[start[0]:end[0]]
    for i in range(1,10) :
        start.append(end[i-1]+7)
        end.append(start[i]+19)
        team_score[i] = score[start[i]:end[i]]
    
    #팀 이름
    team=[i.get_text() for i in soup.select('td')] ; team=team[27::26][:10]

    for i in range(len(team)) :
        if team[i]=='KIA' : team[i]='HT'
        if team[i]=='롯데' : team[i]='LT'
        if team[i]=='두산' : team[i]='OB'
        if team[i]=='키움' : team[i]='WO'
        if team[i]=='삼성' : team[i]='SS'
        if team[i]=='한화' : team[i]='HH'

    #지표들
    var=[i.get_text() for i in soup.select('a')] ; var=var[83:83+18] ; var.insert(0,'AVG')

    for i in range(len(var)) :
        if var[i]=='타석' : var[i]='PA'
        if var[i]=='타수' : var[i]='AB'
        if var[i]=='득점' : var[i]='RUN'
        if var[i]=='안타' : var[i]='HIT'
        if var[i]=='2타' : var[i]='H2'
        if var[i]=='3타' : var[i]='H3'
        if var[i]=='홈런' : var[i]='HR'
        if var[i]=='루타' : var[i]='delete'
        if var[i]=='타점' : var[i]='RBI'
        if var[i]=='도루' : var[i]='SB'
        if var[i]=='도실' : var[i]='CS'
        if var[i]=='볼넷' : var[i]='BB'
        if var[i]=='사구' : var[i]='HP'
        if var[i]=='고4' : var[i]='IB'
        if var[i]=='삼진' : var[i]='KK'
        if var[i]=='병살' : var[i]='GD'
        if var[i]=='희타' : var[i]='SH'
        if var[i]=='희비' : var[i]='SF'
        
    hitter_2020 = pd.concat([pd.Series(var), pd.Series(team_score[0])], axis=1)

    for i in range(1,10) :
        hitter_2020 = pd.concat([hitter_2020, pd.Series(team_score[i])], axis=1)
    
    hitter_2020.index = hitter_2020.iloc[:,0]
    hitter_2020 = hitter_2020.iloc[:,1:]
    hitter_2020.columns = team
    hitter_2020 = hitter_2020.T
    # train set과 형식을 맞추기 위한 열 추가 및 삭제
    hitter_2020 = hitter_2020.drop('delete', axis=1)
    hitter_2020['ERR'] = 0
    hitter_2020['LOB'] = 0
    
    return hitter_2020
```

```python
home_hitter_2020 = get_hitter('http://www.statiz.co.kr/stat.php?mid=stat&re=0&ys=2020&ye=2020&se=0&te=&tm=&ty=0&qu=auto&po=0&as=&ae=&hi=&un=&pl=&da=1&o1=WAR_ALL_ADJ&o2=TPA&de=1&lr=5&tr=&cv=&ml=1&sn=30&pa=0&cn=&si=999&si_it=&si_wd=&si_tm=&si_ha=1&si_te=&si_st=&si_as=&si_or=&si_ty=&si_pl=&si_in=&si_on=&si_um=&si_oc=&si_bs=&si_sc=&si_cnt=&si_aft=&si_li=')
away_hitter_2020 = get_hitter('http://www.statiz.co.kr/stat.php?mid=stat&re=0&ys=2020&ye=2020&se=0&te=&tm=&ty=0&qu=auto&po=0&as=&ae=&hi=&un=&pl=&da=15&o1=AVG&o2=TPA&de=1&lr=5&tr=&cv=&ml=1&sn=30&pa=0&cn=&si=999&si_it=&si_wd=&si_tm=&si_ha=2&si_te=&si_st=&si_as=&si_or=&si_ty=&si_pl=&si_in=&si_on=&si_um=&si_oc=&si_bs=&si_sc=&si_cnt=&si_aft=&si_li=')
```

```python
home_hitter_2020.iloc[:,1:].to_csv('hitter_home_test_x.csv')
away_hitter_2020.iloc[:,1:].to_csv('hitter_away_test_x.csv')
```
